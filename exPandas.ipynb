### loc
레이블을 사용하여 행에 접근하는 메서드.
소괄호가 아닌 대괄호로 접근한다.

```python
# 모든 행의 2014년 데이터
df.loc[:,'2014년']
```

```python
# 5~10행의 201년 데이터
df.loc[5:10, '2014년']   
# numpy는 종료조건이 미만이었지만,
# pandas의 loc는 종료조건이 이하이다.
```

```python
df
```

```python
# 6~9 행에서 기관명 2014년과 2015년 데이터만 가져와서
# 데이터 프레임에 저장
# shape(4, 3)
df1 = df.loc[6:9, ['기관명', '2014년','2015년']]
df1.shape
```

```python
df1 = df.loc[6:9, '2014년':'2016년']
df1
```
